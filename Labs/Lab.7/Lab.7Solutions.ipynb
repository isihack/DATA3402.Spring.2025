```python
#PART A

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the SUSY Dataset


columns = [
    'label', 'lep1_pT', 'lep1_eta', 'lep1_phi', 'lep2_pT', 'lep2_eta', 'lep2_phi',
    'MET', 'MET_phi', 'jet1_pT', 'jet1_eta', 'jet1_phi', 'jet1_btag',
    'jet2_pT', 'jet2_eta', 'jet2_phi', 'jet2_btag', 'jet3_pT', 'jet3_eta',
    'jet3_phi', 'jet3_btag'
]


df_susy = pd.read_csv(
    'SUSY.csv.gz',  
    header=None, 
    names=columns
)

# Define Feature Groups
low_level_features = [
    'lep1_pT', 'lep1_eta', 'lep1_phi', 
    'lep2_pT', 'lep2_eta', 'lep2_phi', 
    'MET', 'MET_phi'
]

high_level_features = [
    'MET', 'lep1_pT', 'lep2_pT', 'jet1_pT'
]


# Define Pair Plot Function

def create_pair_plots(data, features, label_col='label', sample_frac=0.01):
    # Subsample data to speed up plotting
    signal = data[data[label_col] == 1].sample(frac=sample_frac)
    bkg = data[data[label_col] == 0].sample(frac=sample_frac)
    
    n = len(features)
    fig, axes = plt.subplots(n, n, figsize=(15, 15))
    
    for i, feat_i in enumerate(features):
        for j, feat_j in enumerate(features):
            ax = axes[i, j]
            if i == j:
                # Diagonal: Histograms
                ax.hist(signal[feat_i], bins=50, alpha=0.5, density=True, label='Signal')
                ax.hist(bkg[feat_i], bins=50, alpha=0.5, density=True, label='Background')
                ax.set_title(feat_i, fontsize=8)
            else:
                # Off-diagonal: Scatterplots (subsampled)
                ax.scatter(bkg[feat_j], bkg[feat_i], s=1, alpha=0.1, c='red', label='Background')
                ax.scatter(signal[feat_j], signal[feat_i], s=1, alpha=0.1, c='blue', label='Signal')
                ax.set_xticks([])
                ax.set_yticks([])
    
    plt.tight_layout()
    plt.legend(
        handles=[
            plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=5),
            plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=5)
        ],
        labels=['Signal', 'Background'], loc='upper right'
    )
    return fig


#  Generate Plots

# Plot low-level features
fig_low = create_pair_plots(df_susy, low_level_features, sample_frac=0.01)
fig_low.suptitle('Low-Level Features', fontsize=16)
plt.show()

# Plot high-level
fig_high = create_pair_plots(df_susy, high_level_features, sample_frac=0.01)
fig_high.suptitle('High-Level Features (Illustrative)', fontsize=16)
plt.show()
```

```python
#PART B

import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  

def create_pair_plots_fast(data, features, label_col='label', sample_frac=0.01, bins=20):
  
    signal = data[data[label_col] == 1].sample(frac=sample_frac)
    bkg = data[data[label_col] == 0].sample(frac=sample_frac)
    
    n = len(features)
    fig, axes = plt.subplots(n, n, figsize=(15, 15))
    
    for i, feat_i in enumerate(features):
        for j, feat_j in enumerate(features):
            ax = axes[i, j]
            if i == j:
                # Diagonal: Histograms
                ax.hist(signal[feat_i], bins=50, alpha=0.5, density=True, label='Signal')
                ax.hist(bkg[feat_i], bins=50, alpha=0.5, density=True, label='Background')
                ax.set_title(feat_i, fontsize=8)
            else:
                
                # Compute histograms for background 
                counts_bkg, xedges, yedges = np.histogram2d(
                    bkg[feat_j], bkg[feat_i], bins=bins, density=True
                )
                # Compute histograms for signal 
                counts_sig, _, _ = np.histogram2d(
                    signal[feat_j], signal[feat_i], bins=[xedges, yedges], density=True
                )
                
                # Plot background
                ax.imshow(
                    counts_bkg.T,
                    cmap='Reds',
                    origin='lower',
                    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
                    alpha=0.5,
                    aspect='auto'
                )
                # Plot signal
                ax.imshow(
                    counts_sig.T,
                    cmap='Blues',
                    origin='lower',
                    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
                    alpha=0.5,
                    aspect='auto'
                )
                ax.set_xticks([])
                ax.set_yticks([])
    
    plt.tight_layout()
    plt.legend(
        handles=[
            plt.Rectangle((0,0), 1, 1, fc='blue', alpha=0.5),
            plt.Rectangle((0,0), 1, 1, fc='red', alpha=0.5)
        ],
        labels=['Signal', 'Background'],
        loc='upper right'
    )
    return fig
```

```python


import numpy as np
import pandas as pd
from IPython.display import HTML, display
from tabulate import tabulate  # Correct import


# Define column names
columns = [
    'label', 'lep1_pT', 'lep1_eta', 'lep1_phi', 'lep2_pT', 'lep2_eta', 'lep2_phi',
    'MET', 'MET_phi', 'jet1_pT', 'jet1_eta', 'jet1_phi', 'jet1_btag',
    'jet2_pT', 'jet2_eta', 'jet2_phi', 'jet2_btag', 'jet3_pT', 'jet3_eta',
    'jet3_phi', 'jet3_btag'
]

# Load dataset (sampling for faster testing)
df_susy = pd.read_csv('SUSY.csv.gz', header=None, names=columns)



# Compute High-Level Features (Simplified)

# Physics-inspired features
df_susy['MT2'] = np.sqrt(
    2 * df_susy['lep1_pT'] * df_susy['lep2_pT'] *
    (1 - np.cos(df_susy['lep1_phi'] - df_susy['lep2_phi']))
)

df_susy['ET_rel'] = np.where(
    abs(df_susy['MET_phi']) < np.pi / 2,
    df_susy['MET'] * np.sin(df_susy['MET_phi']),
    df_susy['MET']
)

df_susy['Razor_R'] = df_susy['MET'] / (df_susy['lep1_pT'] + df_susy['lep2_pT'])


# Define Analysis Functions

def compute_matrices(data, features):
    """Compute covariance and correlation matrices."""
    X = data[features].values.T
    return np.cov(X), np.corrcoef(X)

def create_table(matrix, features, title, decimals=3):
    """Generate formatted HTML table."""
    rounded = np.round(matrix, decimals)
    table_data = [[features[i]] + list(row) for i, row in enumerate(rounded)]
    
    # Generate the table without caption
    table_html = tabulate(
        table_data,
        headers=["Feature"] + features,
        tablefmt='html',
        stralign='center',
        numalign='center'
    )
    
    # Combine title and table
    full_html = f"<h3>{title}</h3>" + table_html
    return HTML(full_html)

def analyze_features(data, low_features, high_features, decimals=3):
    """Complete analysis pipeline."""
    # Compute matrices
    low_cov, low_corr = compute_matrices(data, low_features)
    high_cov, high_corr = compute_matrices(data, high_features)

    # Display results
    display(HTML("<h2 style='color:#1f77b4'>Low-Level Features Analysis</h2>"))
    display(create_table(low_cov, low_features, "Covariance Matrix", decimals))
    display(create_table(low_corr, low_features, "Correlation Matrix", decimals))

    display(HTML("<h2 style='color:#ff7f0e'>High-Level Features Analysis</h2>"))
    display(create_table(high_cov, high_features, "Covariance Matrix", decimals))
    display(create_table(high_corr, high_features, "Correlation Matrix", decimals))


#Define Features and Run Analysis

# Low-level features (raw detector measurements)
low_level_features = [
    'lep1_pT', 'lep1_eta', 'lep1_phi',
    'lep2_pT', 'lep2_eta', 'lep2_phi',
    'MET', 'MET_phi'
]

# High-level features (physics-inspired)
high_level_features = ['MT2', 'ET_rel', 'Razor_R']

# Execute analysis
analyze_features(df_susy, low_level_features, high_level_features)
```

### Exercise 5.3
Assume 3 different scenarios corresponding to different numbers of signal and background events expected in data:

1. Expect $N_S=10$, $N_B=100$.
1. Expect $N_S=100$, $N_B=1000$.
1. Expect $N_S=1000$, $N_B=10000$.

Plot the significance ($\sigma_{S'}$) for each observable as function of $x_c$ for each scenario, where 

$\sigma_{S'}= \frac{N'_S}{\sqrt{N'_S+N'_B}}$

and $N'_{S,B} = \epsilon_{S,B}(x_c) * N_{S,B}$.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# =============================================
# Load Data
# =============================================

# Define column names
VarNames = ["signal", "MET_rel", "M_R", "MT2", "axial_MET"]  # Add all columns as needed

# Check if file exists
filename = "SUSY-small.csv"
if not os.path.exists(filename):
    raise FileNotFoundError(f"Error: {filename} not found in {os.getcwd()}")

# Load dataset
df = pd.read_csv(filename, names=VarNames, dtype='float64')

# Ensure the 'signal' column exists
if "signal" not in df.columns:
    raise KeyError("Column 'signal' not found in dataset. Check file structure.")

# Split signal and background data
df_sig = df[df["signal"] == 1]
df_bkg = df[df["signal"] == 0]

# Example variables to analyze
variables = ["MET_rel", "M_R", "MT2", "axial_MET"]

# =============================================
# Exercise 5.1: Determine Optimal Cut Direction
# =============================================
def get_cut_direction(sig_data, bkg_data, var):
    """Determine optimal cut direction for a variable."""
    sig_mean = np.mean(sig_data[var])
    bkg_mean = np.mean(bkg_data[var])
    return "x > x_c" if sig_mean > bkg_mean else "x < x_c"

# Store cut directions for each variable
cut_directions = {var: get_cut_direction(df_sig, df_bkg, var) for var in variables}

# =============================================
# Exercise 5.2: Compute TPR/FPR vs. x_c
# =============================================
def compute_efficiencies(sig_data, bkg_data, var, cut_direction, n_steps=1000):
    """Compute TPR (signal efficiency) and FPR (background efficiency)."""
    x_min = min(sig_data[var].min(), bkg_data[var].min())
    x_max = max(sig_data[var].max(), bkg_data[var].max())
    x_c_values = np.linspace(x_min, x_max, n_steps)
    
    tpr, fpr = [], []
    for xc in x_c_values:
        if cut_direction == "x > x_c":
            tpr.append(np.mean(sig_data[var] > xc))
            fpr.append(np.mean(bkg_data[var] > xc))
        else:
            tpr.append(np.mean(sig_data[var] < xc))
            fpr.append(np.mean(bkg_data[var] < xc))
    
    return x_c_values, np.array(tpr), np.array(fpr)

# =============================================
# Exercise 5.3: Compute Significance (σ)
# =============================================
def compute_significance(tpr, fpr, N_S, N_B):
    """Calculate discovery significance σ = N'_S / sqrt(N'_S + N'_B)."""
    N_S_prime = tpr * N_S
    N_B_prime = fpr * N_B
    return N_S_prime / np.sqrt(N_S_prime + N_B_prime + 1e-10)  # Avoid division by zero

# Define scenarios (N_S, N_B)
scenarios = [
    (10, 100),
    (100, 1000),
    (1000, 10000),
    (10000, 100000)
]

# =============================================
# Plot Results for All Variables
# =============================================
for var in variables:
    # Get efficiencies
    x_c, tpr, fpr = compute_efficiencies(df_sig, df_bkg, var, cut_directions[var])
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))
    plt.suptitle(f"Analysis for {var}", y=1.02)
    
    # -----------------------------------------
    # Plot 1: TPR/FPR vs. x_c (Exercise 5.2)
    # -----------------------------------------
    ax1.plot(x_c, tpr, label="Signal Efficiency (TPR)", color="blue")
    ax1.plot(x_c, fpr, label="Background Efficiency (FPR)", color="red")
    ax1.set_xlabel(f"{var} Cut Value ($x_c$)", fontsize=12)
    ax1.set_ylabel("Efficiency", fontsize=12)
    ax1.legend()
    ax1.grid(True, linestyle="--", alpha=0.7)
    
    # -----------------------------------------
    # Plot 2: Significance vs. x_c (Exercise 5.3)
    # -----------------------------------------
    for N_S, N_B in scenarios:
        significance = compute_significance(tpr, fpr, N_S, N_B)
        ax2.plot(x_c, significance, label=f"$N_S$={N_S}, $N_B$={N_B}")
    
    ax2.set_xlabel(f"{var} Cut Value ($x_c$)", fontsize=12)
    ax2.set_ylabel(r"Significance ($\sigma_{S'}$)", fontsize=12)  # Fixed escape sequence
    ax2.legend()
    ax2.grid(True, linestyle="--", alpha=0.7)
    
    plt.tight_layout()
    plt.show()

# =============================================
# Print Optimal Cut Directions (Exercise 5.1)
# =============================================
print("\nOptimal Cut Directions:")
for var, direction in cut_directions.items():
    print(f"{var}: {direction}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier


# STEP 1: LOAD AND PREPARE DATA

def load_data(filepath, nrows=None):
    cols = ["signal"] + [f"feat_{i}" for i in range(1, 19)]
    df = pd.read_csv(filepath, names=cols, nrows=nrows)
    
    # Map features to physics quantities 
    df = df.rename(columns={
        "feat_1": "lep1_pT", "feat_2": "lep1_eta", "feat_3": "lep1_phi",
        "feat_4": "lep2_pT", "feat_5": "lep2_eta", "feat_6": "lep2_phi",
        "feat_7": "MET", "feat_8": "MET_phi",
        "feat_9": "HT", "feat_10": "MT",
        "feat_11": "nJets", "feat_12": "jet1_pT"
    })
    return df


#HIGH-LEVEL FEATURES

def add_high_level_features(df):
    """Calculate MT2 (stransverse mass) and relative missing energy."""
    # Simplified MT2 approximation
    df["MT2"] = np.sqrt(2 * df["lep1_pT"] * df["MET"] * (1 - np.cos(df["lep1_phi"] - df["MET_phi"])))
    df["ET_rel"] = df["MET"] / np.sqrt(df["HT"])
    return df


 #CUT-FLOW ANALYSIS

def cut_flow_table(signal_df, bkg_df, cuts):
    """Generate cut-flow table with metrics."""
    s_counts = [len(signal_df)]
    b_counts = [len(bkg_df)]
    
    for var, condition, _ in cuts:
        signal_df = signal_df.query(condition)
        bkg_df = bkg_df.query(condition)
        s_counts.append(len(signal_df))
        b_counts.append(len(bkg_df))
    
    s_counts = np.array(s_counts)
    b_counts = np.array(b_counts)
    
    with np.errstate(divide='ignore', invalid='ignore'):
        soverb = s_counts / np.sqrt(b_counts)
        s_div_b = s_counts / b_counts
        significance = s_counts / np.sqrt(s_counts + b_counts)
    
    return pd.DataFrame({
        "Cut Step": ["Initial"] + [desc for _, _, desc in cuts],
        "Signal (S)": s_counts,
        "Background (B)": b_counts,
        "S/√B": np.round(soverb, 2),
        "S/B": np.round(s_div_b, 3),
        "Significance (Z)": np.round(significance, 2)
    })

#  CORRELATION ANALYSIS

def plot_correlations(df, title):
    """Plot correlation matrix for key variables."""
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[["MET", "MT", "nJets", "HT", "MT2", "ET_rel"]].corr(), 
                annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title(title)
    plt.show()

def reoptimize_cut(signal_df, bkg_df, fixed_var, fixed_cut, optimize_var, cut_range):
    """Reoptimize a cut after fixing another correlated variable."""
    sig_cut = signal_df.query(f"{fixed_var} > {fixed_cut}")
    bkg_cut = bkg_df.query(f"{fixed_var} > {fixed_cut}")
    
    cut_values = np.linspace(*cut_range, 50)
    significances = []
    
    for cut in cut_values:
        s = len(sig_cut.query(f"{optimize_var} > {cut}"))
        b = len(bkg_cut.query(f"{optimize_var} > {cut}"))
        z = s / np.sqrt(s + b) if (s + b) > 0 else 0
        significances.append(z)
    
    optimal_cut = cut_values[np.argmax(significances)]
    
    plt.plot(cut_values, significances)
    plt.axvline(optimal_cut, color='r', linestyle='--')
    plt.xlabel(f"Cut on {optimize_var}")
    plt.ylabel("Significance (Z)")
    plt.title(f"Optimizing {optimize_var} after {fixed_var} > {fixed_cut}")
    plt.show()
    
    return optimal_cut


# FEATURE IMPORTANCE

def plot_feature_importance(X, y):
    """Plot feature importance using RandomForest."""
    clf = RandomForestClassifier(n_estimators=100)
    clf.fit(X, y)
    
    plt.bar(X.columns, clf.feature_importances_)
    plt.xticks(rotation=45)
    plt.ylabel("Importance")
    plt.title("Feature Importance for SUSY vs. SM")
    plt.show()


# MAIN EXECUTION

if __name__ == "__main__":
    # Load data
    df = load_data("SUSY.csv", nrows=1e5)  # Use subset for quick testing
    df = add_high_level_features(df)
    
    # Split into signal and background
    signal = df[df["signal"] == 1]
    bkg = df[df["signal"] == 0]
    
    # Define cut scenarios
    cuts_lowlevel = [
        ("MET", "MET > 50", "MET > 50 GeV"),
        ("MT", "MT > 40", "MT > 40 GeV"),
        ("nJets", "nJets <= 2", "nJets ≤ 2")
    ]
    
    cuts_highlevel = [
        ("MT2", "MT2 > 100", "MT2 > 100 GeV"),
        ("ET_rel", "ET_rel > 5", "ET_rel > 5 GeV^0.5"),
        ("nJets", "nJets >= 2", "nJets ≥ 2")
    ]
    
    # Generate cut-flow tables
    print("="*50)
    print("Low-Level Features Cut-Flow Table:")
    print(cut_flow_table(signal, bkg, cuts_lowlevel))
    
    print("\n" + "="*50)
    print("High-Level Features Cut-Flow Table:")
    print(cut_flow_table(signal, bkg, cuts_highlevel))
    
    # Correlation analysis
    print("\n" + "="*50)
    print("Correlation Analysis")
    plot_correlations(signal, "Signal Correlation Matrix")
    plot_correlations(bkg, "Background Correlation Matrix")
    
    # Reoptimize MET vs MT cuts
    optimal_MT = reoptimize_cut(signal, bkg, "MET", 50, "MT", (20, 100))
    print(f"\nOptimal MT cut after MET > 50 GeV: {optimal_MT:.1f}")
    
    # Feature importance
    print("\n" + "="*50)
    print("Feature Importance Ranking")
    X = df[["lep1_pT", "MET", "HT", "MT", "MT2", "ET_rel"]]
    y = df["signal"]
    plot_feature_importance(X, y)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_roc_curves(df, observables, conditions=None, ax=None, title=None):
    """
    Plot ROC curves for multiple observables with AUC in legend.
    
    Parameters:
        df (pd.DataFrame): The dataset containing signal labels and observables.
        observables (list): List of column names to compute ROC curves for.
        conditions (str, optional): Query conditions to filter the dataset.
        ax (matplotlib.axes, optional): Axes object for the plot.
        title (str, optional): Title of the plot.
        
    Returns:
        ax: The plot's Axes object.
    """
    if conditions:
        df = df.query(conditions)
        if df.empty:
            print(f"Warning: No data left after applying conditions: {conditions}")
            return None
    
    y_true = df['signal']
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 6))
    
    for obs in observables:
        if obs not in df.columns:
            print(f"Warning: {obs} not found in DataFrame. Skipping.")
            continue
        
        y_score = df[obs]
        fpr, tpr, _ = roc_curve(y_true, y_score)
        roc_auc = auc(fpr, tpr)
        
        ax.plot(fpr, tpr, lw=2, label=f'{obs} (AUC = {roc_auc:.3f})')
    
    ax.plot([0, 1], [0, 1], 'k--', lw=1)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(title or 'ROC Curves Comparison')
    ax.legend(loc="lower right")
    return ax

def apply_selections(df, selection_steps):
    """
    Applies a sequence of selections to the dataset.

    Parameters:
        df (pd.DataFrame): The dataset.
        selection_steps (list of str): List of filtering conditions.
    
    Returns:
        list: List of DataFrames at each step.
    """
    df_steps = [df]
    for condition in selection_steps:
        df = df.query(condition)
        if df.empty:
            print(f"Warning: No data left after selection: {condition}")
            break
        df_steps.append(df)
    return df_steps

def main():
    print("Loading dataset...")
    try:
        df = pd.read_csv("SUSY.csv", nrows=100000, header=None)
    except FileNotFoundError:
        raise FileNotFoundError("SUSY.csv not found. Please ensure the file exists in your working directory.")

    print("Renaming columns...")
    df.columns = [
        'signal', 'lep1_pT', 'lep1_eta', 'lep1_phi',
        'lep2_pT', 'lep2_eta', 'lep2_phi',
        'MET', 'MET_phi',
        'HT', 'MT',
        'nJets', 'jet1_pT', 'feat_13', 'feat_14',
        'feat_15', 'feat_16', 'feat_17', 'feat_18'
    ]

    # Compute high-level features
    if {'lep1_pT', 'MET', 'lep1_phi', 'MET_phi', 'HT'}.issubset(df.columns):
        print("Computing high-level features...")
        df['MT2'] = np.sqrt(2 * df['lep1_pT'] * df['MET'] * 
                            (1 - np.cos(df['lep1_phi'] - df['MET_phi'])))
        df['ET_rel'] = df['MET'] / np.sqrt(df['HT'])
    else:
        raise KeyError("Missing required columns for feature computation!")

    # Verify signal column exists
    if 'signal' not in df.columns:
        raise ValueError("DataFrame must contain 'signal' column with 0 (background) and 1 (signal) values")
    
    print("Generating ROC comparison plot...")
    top_obs = ['MT2', 'ET_rel', 'MET']
    plt.figure()
    plot_roc_curves(df, top_obs, title="ROC Comparison of Top Observables")
    plt.show()  # Display plot
    plt.savefig('roc_comparison.png', bbox_inches='tight')
    print("ROC comparison plot saved as roc_comparison.png.")
    plt.close()

    # Exercise 7.3: Apply successive selections
    print("Applying successive selections...")
    selection_steps = [
        "lep1_pT > 20",
        "lep2_pT > 15",
        "MET > 30",
        "HT > 100"
    ]
    
    df_steps = apply_selections(df, selection_steps)

    print("Plotting ROC curves after successive selections...")
    fig, ax = plt.subplots(figsize=(8, 6))
    for i, df_step in enumerate(df_steps):
        plot_roc_curves(df_step, ['MT2'], ax=ax, title="ROC Curves After Successive Selections")
    plt.show()  # Display plot
    plt.savefig('roc_successive_selections.png', bbox_inches='tight')
    print("ROC curves for successive selections saved as roc_successive_selections.png.")
    plt.close()

    # Exercise 7.4: Test effect of selection order
    print("Testing effect of selection order...")
    alt_selection_steps = [
        "HT > 100",
        "MET > 30",
        "lep1_pT > 20",
        "lep2_pT > 15"
    ]
    
    df_alt_steps = apply_selections(df, alt_selection_steps)

    fig, ax = plt.subplots(figsize=(8, 6))
    for i, df_step in enumerate(df_alt_steps):
        plot_roc_curves(df_step, ['MT2'], ax=ax, title="Effect of Selection Order on ROC Curves")
    plt.show()  # Display plot
    plt.savefig('roc_selection_order.png', bbox_inches='tight')
    print("ROC curves for selection order saved as roc_selection_order.png.")
    plt.close()

    print("ROC analysis completed successfully!")

if __name__ == "__main__":
    main()
```

## Exercise 8: Linear Discriminant

### Exercise 8.1

Using numpy, compute the between-class $\bf{S}_B$ and within-class $\bf{S}_W$ covariance matrices defined as:

$$
\bf{S}_B = (\bf{m_2}-\bf{m_1})(\bf{m_2}-\bf{m_1})^T \\
$$
$$
\bf{S}_W = \sum_{i=1,2} \sum_{n=1}^{l_i} (\bf{x}_n^i - \bf{m}_i) (\bf{x}_n^i - \bf{m}_i)^T
$$

where $\bf{m_i}$ are the vectors containing the means for category 1 and 2, here defined as signal and background. Here $\bf{x}_n^i$ is the vector containing the observables for the $n$th example event in category $i$.

### Exercise 8.1

Compute the linear coefficients $\bf{w} = \bf{S_W}^{-1}(\bf{m_2}-\bf{m_1})$. Compare the histogram of the distribution of $F_n^i=\bf{w}^T\bf{x}_n^i$ for the two categories.

### Exercise 8.1

Draw the ROC curve for $F_n$.

What is the maximal significance you can obtain in the scenarios in exercise 5? 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import StandardScaler
import os

# =============================================
# Load Data with Validation
# =============================================
def load_and_validate_data():
    """Load and validate SUSY data with diagnostics"""
    filename = "SUSY.csv"
    print(f"Looking for {filename} in {os.getcwd()}...")

    if not os.path.exists(filename):
        print("Error: File not found. Available files:")
        print(os.listdir(os.getcwd()))
        raise FileNotFoundError(f"{filename} not found in directory.")

    print("Loading data...")
    VarNames = ["signal", "MET_rel", "M_R", "MT2", "axial_MET"]

    try:
        df = pd.read_csv(filename, names=VarNames)
        print(f"Data loaded successfully. Shape: {df.shape}")

        # Ensure signal is strictly binary
        if not set(df['signal'].unique()).issubset({0, 1}):
            print("Warning: Converting signal column to binary (0/1)...")
            df['signal'] = (df['signal'] > 0.5).astype(int)

        print("Signal distribution:\n", df['signal'].value_counts())
        return df

    except Exception as e:
        print(f"Data loading failed: {e}")
        raise

# =============================================
# Compute LDA Components
# =============================================
def compute_lda_components(df, features):
    """Compute LDA components with validation and feature scaling"""
    print("\nComputing LDA components...")
    print("Using features:", features)

    # Validate features
    missing = [f for f in features if f not in df.columns]
    if missing:
        raise KeyError(f"Missing features: {missing}")

    # Scale features
    scaler = StandardScaler()
    df[features] = scaler.fit_transform(df[features])

    # Separate signal & background
    df_sig = df[df['signal'] == 1][features]
    df_bkg = df[df['signal'] == 0][features]

    print(f"Signal events: {len(df_sig)}, Background events: {len(df_bkg)}")

    if len(df_sig) == 0 or len(df_bkg) == 0:
        raise ValueError("Error: No signal or background events found.")

    # Compute class means
    m1 = np.mean(df_sig.values, axis=0)
    m2 = np.mean(df_bkg.values, axis=0)

    # Compute between-class covariance
    diff = (m2 - m1).reshape(-1, 1)
    S_B = np.dot(diff, diff.T)

    # Compute within-class covariance
    def compute_scatter(data):
        centered = data - np.mean(data, axis=0)
        return np.dot(centered.T, centered)

    S_W = compute_scatter(df_sig) + compute_scatter(df_bkg)
    print("Covariance matrices computed.")

    return S_B, S_W, m1, m2

# =============================================
# Compute LDA Discriminant
# =============================================
def compute_linear_discriminant(S_W, m1, m2):
    """Compute LDA weights with stability checks"""
    print("\nComputing LDA weights...")

    try:
        # Use pseudo-inverse for numerical stability
        w = np.linalg.pinv(S_W) @ (m2 - m1)
        w /= np.linalg.norm(w)  # Normalize
        print("LDA weights computed successfully.")
        return w
    except np.linalg.LinAlgError:
        raise ValueError("Matrix inversion failed. Check for singularity.")

# =============================================
# Main Execution
# =============================================
def main():
    try:
        # Load data
        df = load_and_validate_data()
        features = ["MET_rel", "M_R", "MT2", "axial_MET"]

        # Compute LDA components
        S_B, S_W, m1, m2 = compute_lda_components(df, features)

        # Compute discriminant
        w = compute_linear_discriminant(S_W, m1, m2)

        print("\nLDA Weights:")
        for name, weight in zip(features, w):
            print(f"{name}: {weight:.4f}")

        # Project data
        F = df[features].values @ w
        print(f"\nProjection range: {F.min():.2f} to {F.max():.2f}")

        # Compute ROC Curve
        plt.figure(figsize=(12, 5))

        # Histogram plot
        plt.subplot(121)
        plt.hist(F[df['signal'] == 1], bins=50, alpha=0.5, label='Signal')
        plt.hist(F[df['signal'] == 0], bins=50, alpha=0.5, label='Background')
        plt.xlabel("LDA Projection")
        plt.ylabel("Counts")
        plt.legend()

        # ROC Curve plot
        plt.subplot(122)
        fpr, tpr, _ = roc_curve(df['signal'], F)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.legend()

        plt.tight_layout()
        plt.savefig('lda_results.png')
        print("\nSaved plot: lda_results.png")

        # Compute significance
        print("\nComputing maximal significance...")
        scenarios = [(10, 100), (100, 1000), (1000, 10000)]
        results = {}

        best_z = 0
        best_t = 0
        thresholds = np.linspace(F.min(), F.max(), num=100)

        for N_S, N_B in scenarios:
            for t in thresholds:
                pred = (F > t).astype(int)
                TP = np.sum((pred == 1) & (df['signal'] == 1))
                FP = np.sum((pred == 1) & (df['signal'] == 0))

                # Compute significance metric
                Z = TP / np.sqrt(TP + FP + 1e-10)  # Avoid divide-by-zero
                
                if Z > best_z:
                    best_z = Z
                    best_t = t

            results[f"N_S={N_S}, N_B={N_B}"] = best_z

        print("\nOptimal Significance Threshold:")
        print(f"Best threshold: {best_t:.3f}, Max Z: {best_z:.2f}σ")

        print("\nSignificance results:")
        for scenario, z in results.items():
            print(f"{scenario}: {z:.2f}σ")

    except Exception as e:
        print("\nError encountered:", str(e))
        plt.close('all')

if __name__ == "__main__":
    print("=== Running SUSY LDA Analysis ===")
    main()
    print("\nAnalysis complete!")
```
